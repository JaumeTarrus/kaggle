```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.deterministic import DeterministicProcess, CalendarFourier
from sklearn.linear_model import LinearRegression
```

```python
dtype = {
    'store_nbr': 'category',
    'family': 'category',
    'sales': 'float32',
    'onpromotion': 'uint64',
}

train = pd.read_csv('train.csv', parse_dates = ['date'], dtype=dtype)
train['date'] = train.date.dt.to_period('D')
train = train.set_index(['store_nbr', 'family', 'date']).sort_index()


test = pd.read_csv('test.csv', parse_dates = ['date'], dtype=dtype)
test['date'] = test.date.dt.to_period('D')

oil = pd.read_csv('oil.csv',parse_dates = True, index_col=['date']).to_period('D')
holidays_events = pd.read_csv('holidays_events.csv', parse_dates = True, index_col=['date']).to_period('D')
stores = pd.read_csv('stores.csv')
```

```python
avg_sales = train.groupby('date').mean()['sales']
avg_sales_trend = avg_sales.rolling(window = '7D', min_periods=1, center = True).mean()

start='2016-01-30'
finish='2017-07-30'

avg_sales_trend_slice = avg_sales_trend.loc[start:finish]
index=avg_sales_trend_slice.index

dp = DeterministicProcess(index=index , order=1, drop=True)

y = avg_sales_trend_slice.values
X = dp.in_sample()
X_forecast = dp.out_of_sample(steps=16)


model = LinearRegression()
model.fit(X, y)
y_pred = pd.Series(model.predict(X), index=index)
y_forecast = pd.Series(model.predict(X_forecast), index=X_forecast.index)

fig, ax = plt.subplots(figsize=(9, 3), layout='constrained')

ax.plot(index.to_timestamp(),y)
ax.plot(X_forecast.index.to_timestamp(),avg_sales_trend[X_forecast.index])
ax.plot(y_pred)
ax.plot(y_forecast)
#ax.plot(avg_sales[start:X_forecast.index[-1]],  alpha=0.5)
ax.set(title="Avg Sales", ylabel="Do not actually know the units", xlabel="Dates");
```

```python
oil = pd.read_csv('oil.csv',parse_dates = True, index_col=['date']).to_period('D')
oil = oil['dcoilwtico'].resample('D').fillna(method='pad').fillna(method='bfill')
index = avg_sales_trend.index
index_forecast =oil[index[-1]:].index

y = avg_sales_trend.values
X = oil[avg_sales_trend.index].values.reshape(-1, 1)
X_forecast = oil[index[-1]:].values.reshape(-1, 1)

model = LinearRegression()
model.fit(X, y)
y_pred = pd.Series(model.predict(X), index= index)
y_forecast = pd.Series(model.predict(X_forecast), index=index_forecast)

fig, ax = plt.subplots(figsize=(9, 3), layout='constrained')
ax.plot(avg_sales.index.to_timestamp(),avg_sales.values, alpha=0.5)
ax.plot(avg_sales_trend)
ax.plot(oil.index.to_timestamp(), oil.values, "r-")
ax.plot(y_pred)
ax.plot(y_forecast)
ax.set(title="Avg Sales", ylabel="Do not actually know the units", xlabel="Dates");
```

```python
def plot_periodogram(ts, detrend='linear'):
    from scipy.signal import periodogram
    freqencies, spectrum = periodogram(
        ts,
        fs=366,
        detrend=detrend,
        window="boxcar",
        scaling='spectrum',
    )
    plt.figure(figsize=(8, 4), layout='constrained')
    plt.step(freqencies, spectrum, color="purple")
    plt.xscale("log")
    plt.xticks([1, 2, 4, 6, 12, 26, 52, 104], ["Annual (1)","Semiannual (2)","Quarterly (4)","Bimonthly (6)","Monthly (12)",
                                               "Biweekly (26)","Weekly (52)", "Semiweekly (104)"], rotation=90)
    plt.ticklabel_format(axis="y", style="sci", scilimits=(0, 0))
    plt.ylabel("Variance")
    plt.title("Periodogram")
    return plt.show()
```

```python
plot_periodogram(avg_sales)
```

```python
sales = avg_sales[:'2017-08-15']
index = sales.index

fourier_1 = CalendarFourier('M', 2)
fourier_2 = CalendarFourier('A', 2)
#fourier_3 = CalendarFourier('5A', 4)

dp = DeterministicProcess(index = index,
                          order = 1,
                          drop= True,
                          seasonal = True,
                          period = 7,
                          additional_terms=[fourier_2,fourier_1])
# period=365, seasonal= True, additional_terms=[fourier] period=182, fourier=8
y = sales.values
X = dp.in_sample()
X_forecast = dp.out_of_sample(steps=30)

model = LinearRegression()
model.fit(X, y)
y_pred = pd.Series(model.predict(X), index = index)
y_forecast = pd.Series(model.predict(X_forecast), index = X_forecast.index)

fig, ax = plt.subplots(figsize=(9, 3), layout='constrained')

aux = avg_sales[:'2017-08-15']
ax.plot(aux.index.to_timestamp(), aux.values, alpha=0.5)

ax.plot(y_pred)
ax.plot(y_forecast)
ax.set(title="Avg Sales", ylabel="Do not actually know the units", xlabel="Dates");
```

```python
y_ds = pd.Series(y, index = index) -y_pred
y_ds = y_ds.resample('D').fillna(method=None).fillna(0)

hd_n=holidays_events.query("locale in ['National']").loc[:'2017-07-15',['description']]
hd_r=holidays_events.query("locale in ['Regional']").loc[:'2017-07-15',['description']]
hd_l=holidays_events.query("locale == 'Local'").loc[:'2017-07-15',['description']]

fig, ax = plt.subplots(figsize=(9, 3), layout='constrained')

ax.plot(hd_n.index.to_timestamp(), y_ds[hd_n.index], "ro",
        hd_r.index.to_timestamp(), y_ds[hd_r.index], "gs",
        hd_l.index.to_timestamp(), y_ds[hd_l.index], "y*")

ax.plot(y_ds.index.to_timestamp(),y_ds.values)

ax.set_ylim(-200,400)
ax.set(title="Avg Sales Deseasoned", ylabel="Do not actually know the units", xlabel="Dates");
```
