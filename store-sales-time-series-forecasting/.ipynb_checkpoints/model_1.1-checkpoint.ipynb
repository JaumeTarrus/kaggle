```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.deterministic import DeterministicProcess, CalendarFourier
from sklearn.linear_model import LinearRegression
```

```python
dtype = {
    'store_nbr': 'category',
    'family': 'category',
    'sales': 'float32',
    'onpromotion': 'uint64',
}

train = pd.read_csv('train.csv', parse_dates = ['date'], dtype=dtype)
train['date'] = train.date.dt.to_period('D')
train = train.set_index(['store_nbr', 'family', 'date']).sort_index()


test = pd.read_csv('test.csv', parse_dates = ['date'], dtype=dtype)
test['date'] = test.date.dt.to_period('D')

oil = pd.read_csv('oil.csv',parse_dates = True, index_col=['date']).to_period('D')
holidays_events = pd.read_csv('holidays_events.csv', parse_dates = True, index_col=['date']).to_period('D')
```

```python
y = train[['sales']].unstack(['store_nbr', 'family']).loc['2017-01-01':]

index = y.index

fourier_2 = CalendarFourier('M', 4)
fourier_1 = CalendarFourier('A', 2)

dp = DeterministicProcess(index = index,
                          order = 1,
                          drop= True,
                          seasonal = True,
                          period = 7,
                          additional_terms=[fourier_2,fourier_1]
)

X = dp.in_sample()
X_forecast = dp.out_of_sample(steps=16)

model = LinearRegression(fit_intercept=False)
model.fit(X, y)
y_pred = pd.DataFrame(model.predict(X), index = index, columns=y.columns)
y_forecast = pd.DataFrame(model.predict(X_forecast), index = X_forecast.index, columns=y.columns)
y_submit = y_forecast.stack(['store_nbr', 'family']).reset_index()
y_submit = y_submit.join(test[['id']])
y_submit.to_csv('submission.csv', index=False, columns =['id','sales'])
```

```python
holidays_events = pd.read_csv('holidays_events.csv',parse_dates = True, index_col=['date']).to_period('D')
holidays_events = holidays_events.loc['2017-01-01':'2017-08-15']

stores = pd.read_csv('stores.csv', dtype={'store_nbr': 'category', 'city': 'category', 'state': 'category'}, usecols=['store_nbr','city','state'])


strng = ('sales','12','GROCERY I')
aux = y[strng].resample('D').fillna(method=None).fillna(0)
aux_pred = y_pred[strng]
aux_forecast = y_forecast[strng]

city = stores.query('store_nbr  == @strng[1]')['city'].iloc[0]
state = stores.query('store_nbr  == @strng[1]')['state'].iloc[0]
hd_at_store = holidays_events.query('(locale == "National") | (locale == "Regional" and locale_name == @state) | (locale == "Local" and locale_name == @city)')

fig, ax = plt.subplots(figsize=(9, 3), layout='constrained')

ax.plot(aux.index.to_timestamp(), aux.values, alpha=0.5)
ax.plot(hd_at_store.index.to_timestamp(), aux[hd_at_store.index], "ro")
ax.plot(aux_pred)
ax.plot(aux_forecast)
ax.set(title="Avg Sales", ylabel="Do not actually know the units", xlabel="Dates");
```
